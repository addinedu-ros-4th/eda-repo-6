```python
import pandas as pd

transaction = pd.read_csv("../data/sss/transaction21.csv", encoding='utf-8')
transaction.head()
```

```python
school = pd.read_csv("../data/sss/school.csv", encoding='utf-8')
school.head()
```

```python
own_rate = pd.read_csv("../data/sss/own_rate.csv", encoding='utf-8')
own_rate.head()
```

```python
transaction.columns
```

```python
transaction.sort_values(by='구', ascending=True, inplace=True)
transaction.set_index('구', drop=True, inplace=True)

transaction
```

```python
school.sort_values(by='구', ascending=True, inplace=True)
school.set_index('구', drop=True, inplace=True)
school
```

```python
own_rate.rename(columns={'구별':'구'}, inplace=True)
own_rate
```

```python
own_rate.sort_values(by='구', ascending=True, inplace=True)
own_rate.set_index('구', drop=True, inplace=True)
own_rate
```

```python
yjs = transaction.merge(school, left_on='구', right_on='구')
yjs
```

```python
yjs = yjs.merge(own_rate, left_on='구', right_on='구')
yjs
```
