#  서울시 소재 초등학교 수

```python
import pandas as pd

# 2021년_서울_구별 초등학교 개수
seoul_school = pd.read_excel("../data/서울시교육청학교현황(20210401기준).xlsx",
                             sheet_name="초",
                             header=2,
                             engine="openpyxl")
seoul_school.head()
```

```python
seoul_school.columns
```

```python
# 데이터 가공을 위한 인덱스 찾기

start_index = seoul_school.columns.get_loc("국립")
end_index = seoul_school.columns.get_loc("계")

#start_index
#end_index
```

```python
seoul_school = seoul_school.drop(seoul_school.columns[start_index:end_index], axis=1)

seoul_school.head()
```

```python
seoul_school = seoul_school.iloc[:,2:4]
seoul_school
```

```python
seoul_school = seoul_school.drop(seoul_school.loc[seoul_school['소재구']=='소계'].index)
seoul_school
```

```python
seoul_school = seoul_school.drop([0,1,38,39,40,41])
seoul_school
```

```python
seoul_school.rename(columns={'소재구':'구'}, inplace=True)
seoul_school = seoul_school.set_index(keys='구')
seoul_school.to_csv("../data/yjs/03_2021_Seoul_school_yjs.csv")
seoul_school
```

```python
import koreanize_matplotlib
import matplotlib.pyplot as plt

seoul_school = seoul_school.sort_values(by="계", ascending=False)
seoul_school.plot.bar()
plt.ylabel('학생수')
```
