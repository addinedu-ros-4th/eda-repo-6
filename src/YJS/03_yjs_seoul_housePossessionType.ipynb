```python
import pandas as pd

# 2020년 서울시_자치구별_주택점유형태
seoul_own = pd.read_csv("../data/2020년_서울시_자치구별_주택점유형태.csv", encoding='utf-8', header=1)
seoul_own.head()
```

```python
seoul_own = seoul_own.drop([0])
seoul_own
```

```python
#seoul_own_rate.columns[0]
seoul_own = seoul_own.drop(seoul_own.columns[0], axis=1)
seoul_own
```

```python
seoul_own_rate= seoul_own.drop(seoul_own.columns[3:], axis=1)
seoul_own_rate
```

```python
seoul_own_rate["자가율(%)"] = seoul_own_rate["자가"] / seoul_own_rate["합계"] * 100
seoul_own_rate
```

```python
seoul_own_rate.rename(columns={"자치구별(2)":"구별"}, inplace=True)
seoul_own_rate.set_index(keys="구별", inplace=True)
seoul_own_rate
```

```python
seoul_own_rate = seoul_own_rate.sort_values(by="자가율(%)", ascending=False)
seoul_own_rate.to_csv("../data/yjs/04_2021_Seoul_own_yjs.csv")
seoul_own_rate
```

```python
import koreanize_matplotlib
import matplotlib.pyplot as plt

seoul_own_rate.plot.bar(y="자가율(%)")
plt.ylim([25,60])
plt.title("서울시 구별 자가점유율")
```
