```python
import pandas as pd

data = pd.read_csv("../csvfiles/seoul_big_mart.csv",encoding="euc-kr")



```

```python
data.fillna("0",inplace=True)
```

```python
import numpy as np
test_set = data[["인허가일자","폐업일자"]]


def make_new_columns(col_name):
    new_list = []
    for i, ele in enumerate(test_set[col_name]):
        new_ele = ''
        if '-' in ele:
            new_ele = ele.replace('-','')
            new_list.append(new_ele)
        else: 
            new_list.append(ele)
    return new_list

new_ok_list = make_new_columns("인허가일자")        
new_close_list = make_new_columns("폐업일자") 

print(new_ok_list[:30])
print(new_close_list[:30])
```

```python
def make_allow_date(col_name):
    new_date_list =[]
    
    for i, year in enumerate(data[col_name]):
        if year != "0":
            new_date_list.append(int(year[:4]))
        else:
            new_date_list.append(int(year))
    return new_date_list
new_ok_list = make_allow_date("인허가일자")
new_close_list = make_allow_date("폐업일자")
    
```

```python
data["인허가년도"] = new_ok_list
data["폐업년도"] = new_close_list

data.info()
```

```python
change_coordinate_data = data[["좌표정보(X)","좌표정보(Y)"]]
```

```python
change_coordinate_data
```

```python
from pyproj import Proj, transform
import folium

source_proj = Proj(init='epsg:2097')
target_proj = Proj(init='epsg:4326')

# Function to transform a row of coordinates
def transform_coordinates(row):
    lon, lat = transform(source_proj, target_proj, 
                         row['좌표정보(X)'], row['좌표정보(Y)'])
    return pd.Series({'lon': lon, 'lat': lat})

# Apply the transformation to each row
change_coordinate_data[['lon', 'lat']] = change_coordinate_data.apply(transform_coordinates, axis=1)

```

```python
data[["위도","경도"]] = change_coordinate_data[["lon","lat"]]
```
